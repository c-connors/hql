```python
from PIL import Image
#import maze
import random
import numpy as np

def generate_maze(size):
    # Source: http://code.activestate.com/recipes/578356-random-maze-generator/
    # Random Maze Generator using Depth-first Search
    # http://en.wikipedia.org/wiki/Maze_generation_algorithm

    mx = size-2; my = size-2 # width and height of the maze
    maze = np.ones((my, mx))
    dx = [0, 1, 0, -1]; dy = [-1, 0, 1, 0] # 4 directions to move in the maze
    # start the maze from a random cell
    start_x = random.randint(0, mx - 1); start_y = random.randint(0, my - 1)
    cx, cy = 0, 0
    maze[start_y][start_x] = 0; stack = [(start_x, start_y, 0)] # stack element: (x, y, direction)
    while len(stack) > 0:
        (cx, cy, cd) = stack[-1]
        # to prevent zigzags:
        # if changed direction in the last move then cannot change again
        if len(stack) > 2:
            if cd != stack[-2][2]: dirRange = [cd]
            else: dirRange = range(4)
        else: dirRange = range(4)

        # find a new cell to add
        nlst = [] # list of available neighbors
        for i in dirRange:
            nx = cx + dx[i]; ny = cy + dy[i]
            if nx >= 0 and nx < mx and ny >= 0 and ny < my:
                if maze[ny][nx] == 1:
                    ctr = 0 # of occupied neighbors must be 1
                    for j in range(4):
                        ex = nx + dx[j]; ey = ny + dy[j]
                        if ex >= 0 and ex < mx and ey >= 0 and ey < my:
                            if maze[ey][ex] == 0: ctr += 1
                    if ctr == 1: nlst.append(i)

        # if 1 or more neighbors available then randomly select one and move
        if len(nlst) > 0:
            ir = nlst[random.randint(0, len(nlst) - 1)]
            cx += dx[ir]; cy += dy[ir]; maze[cy][cx] = 1
            stack.append((cx, cy, ir))
        else: stack.pop()

    maze_tensor = np.zeros((size, size, 3))
    maze_tensor[:,:,0] = 1
    maze_tensor[1:-1, 1:-1, 0] = maze
    maze_tensor[start_y+1][start_x+1][1] = 1

    # Randomly determine goal position
    while True:
        goal_y = random.randint(0, size-2) + 1
        goal_x = random.randint(0, size-2) + 1

        if maze_tensor[goal_y][goal_x][0] == 0:
            maze_tensor[goal_y][goal_x][2] = 1
            break

    # dist = np.ones((size, size)) * -1
    #queue = [(cy+1, cx+1)]
    #idx = 0
    #while len(queue) > 0:
    #   (y, x) = queue[idx]
    #   idx += 1

    #   for i in range(4):
    #       ny = y + dy[i]; nx = x + dx[i]

def visualize_maze(maze):
    imgx = 500; imgy = 500
    color = [(0, 0, 0), (255, 255, 255)] # RGB colors of the maze
    image = Image.new("RGB", (imgx, imgy))
    pixels = image.load()
    print(maze)
    my, mx = maze.shape[0], maze.shape[1]
    # paint the maze
    for ky in range(imgy):
        for kx in range(imgx):
            if maze[my * ky / imgy][mx * kx / imgx][0] == 1:
                pixels[kx, ky] = (0, 0, 0)
            elif maze[my * ky / imgy][mx * kx / imgx][1] == 1:
                pixels[kx, ky] = (255, 0, 0)
            elif maze[my * ky / imgy][mx * kx / imgx][2] == 1:
                pixels[kx, ky] = (0, 0, 255)
            else:
                pixels[kx, ky] = (255, 255, 255)

    return image

m = generate_maze(10)
visualize_maze(m).show()
```
